```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Load the unemployment data
data = pd.read_csv(r"F:\OASIS\Task2_Unemployment_analysis\archive (1)\Unemployment in India.csv")
```

```python
# Display the columns
print(data.columns)
```

```python
# Ensure the 'Date' column is in datetime format
data[' Date'] = pd.to_datetime(data[' Date'])
```

```python
# Plot the unemployment rate over time
plt.figure(figsize=(10, 6))
plt.plot(data[' Date'], data[' Estimated Unemployment Rate (%)'], marker='o')
plt.xlabel('Date')
plt.ylabel('Unemployment Rate (%)')
plt.title('Unemployment Rate Over Time')
plt.grid(True)
plt.show()
```

```python
# Calculate the average unemployment rate
average_rate = data[' Estimated Unemployment Rate (%)'].mean()
print('Average Unemployment Rate:', average_rate)
```

```python
# Find the date with the highest unemployment rate
highest_rate_date = data[' Date'][data[' Estimated Unemployment Rate (%)'].idxmax()]
print('Date with Highest Unemployment Rate:', highest_rate_date.date())
```

```python
# Extract the year for yearly analysis
data['Year'] = data[' Date'].dt.year
```

```python
# Calculate average unemployment rate by year
yearly_avg_unemployment_rate = data.groupby('Year')[' Estimated Unemployment Rate (%)'].mean().reset_index()
```

```python
# Create a bar chart of average unemployment rates by year
plt.figure(figsize=(10, 6))
plt.bar(yearly_avg_unemployment_rate['Year'], yearly_avg_unemployment_rate[' Estimated Unemployment Rate (%)'])
plt.xlabel('Year')
plt.ylabel('Average Unemployment Rate (%)')
plt.title('Average Unemployment Rates by Year')
plt.xticks(yearly_avg_unemployment_rate['Year'], rotation=45)
plt.grid(True)
plt.show()
```
